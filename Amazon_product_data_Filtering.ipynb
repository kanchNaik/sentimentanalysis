```python
dataset_Electronics = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "raw_review_Electronics", trust_remote_code=True)
```

```python
iter_dataset = iter(dataset_Electronics["full"])
```

```python

# Define the required fields
selected_columns = ["asin", "rating", "title", "text", "timestamp"]
batch_size = 10000  # Number of rows per batch
csv_filename = "amazon_electronics_filtered.csv"

batch_data = []  # Temporary storage for a batch

# Open CSV file in append mode
for i, row in enumerate(iter_dataset):
    # Select only required fields
    filtered_row = {col: row.get(col, None) for col in selected_columns}
    batch_data.append(filtered_row)

    # Process when batch reaches batch_size
    if (i + 1) % batch_size == 0:
        df = pd.DataFrame(batch_data)  # Convert batch to DataFrame
        df.to_csv(csv_filename, mode="a", index=False, header=(i < batch_size))  # Append to CSV
        batch_data = []  # Reset batch

        print(f"Processed {i+1} rows...")

# Save remaining data if any
if batch_data:
    df = pd.DataFrame(batch_data)
    df.to_csv(csv_filename, mode="a", index=False, header=False)
    print("Final batch saved.")

print("Processing complete!")
```

```python
import pandas as pd

# Define file paths
input_file_path = "/content/drive/MyDrive/amazon_electronics_filtered.csv"  # Input file
output_file_path = "/content/drive/MyDrive/amazon_electronics_processed.csv"  # Output file

# Define the chunk size
chunk_size = 100000  # Adjust based on memory availability

# Define required columns (modify if needed)
selected_columns = ["asin", "rating", "title", "text", "timestamp"]

# Initialize a flag to write header only once
first_chunk = True

# Read the file in chunks, process, and append to a new CSV file
for chunk in pd.read_csv(input_file_path, usecols=selected_columns, chunksize=chunk_size):

    # Example Processing: Convert timestamp to datetime and filter records after Jan 1, 2015
    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"], unit="ms")  # Convert from milliseconds
    chunk = chunk[chunk["timestamp"] > "2015-01-01"]  # Filter records

    # Append processed chunk to output CSV
    chunk.to_csv(output_file_path, mode="a", index=False, header=first_chunk)

    # Ensure the header is written only for the first chunk
    first_chunk = False

    print(f"Processed {chunk_size} rows...")

print("Processing complete! ✅")
```
