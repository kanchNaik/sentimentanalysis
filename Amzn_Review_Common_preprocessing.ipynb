```python
input_file = "/content/drive/MyDrive/amazon_electronics_processed.csv"
output_file = "/content/drive/MyDrive/amazon_electronics_cleaned.csv"
```

```python
chunk_size = 10000
```

```python
negation_words = {"not", "no", "never", "n't", "cannot", "hardly", "barely", "scarcely"}
```

```python
def handle_negation(tokens):
    negation_active = False
    processed_tokens = []

    for token in tokens:
        if token in negation_words:  # If a negation word is found, activate negation flag
            negation_active = True
            processed_tokens.append(token)
        elif negation_active and re.match(r"^\w+$", token):  # Apply negation to next words (excluding punctuation)
            processed_tokens.append(token + "_NEG")
        else:
            processed_tokens.append(token)
            negation_active = False  # Reset negation when punctuation or stop word is reached

    return processed_tokens
```

```python
def clean_Text(text):
    if isinstance(text, str):
        text = text.lower()  # Convert to lowercase
        text = re.sub(r"[^a-z\s']", "", text)  # Remove punctuation except apostrophes
        tokens = word_tokenize(text)  # Tokenize the text
        tokens = handle_negation(tokens)
        return tokens  # Return tokenized list

    return []
```

```python
with pd.read_csv(input_file, chunksize=chunk_size) as reader:
    for i, chunk in enumerate(reader):
        # Apply text cleaning and tokenization function
        chunk["title"] = chunk["title"].apply(clean_Text)
        chunk["text"] = chunk["text"].apply(clean_Text)

        # Append processed data to new file
        chunk.to_csv(output_file, mode="a", index=False, header=(i == 0))
        print(f"Processed {chunk_size} rows...")
```
